# 1️⃣ Arithmetic Operations

## What are Arithmetic Operations?

Arithmetic operations are basic mathematical operations such as addition, subtraction, multiplication, division, modulus, floor division, and exponentiation.

Unlike Python lists, NumPy performs these operations **element-wise**.

This means that the operation is performed on corresponding elements of two arrays.

---

## Why is this Important?

Imagine you have a dataset containing one million values.

Using Python loops, performing operations on every value would take much longer.

NumPy performs these operations internally using optimized C code, making them significantly faster.

---

## Real-Life Example

Suppose you have marks of students in two tests.

Test 1

[80, 75, 90]

Test 2

[10, 15, 5]

If bonus marks need to be added:

80 + 10

75 + 15

90 + 5

NumPy performs this automatically without writing loops.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Array A :", a)
print("Array B :", b)
```

## Addition

Addition combines corresponding elements from both arrays.

This is called **Element-wise Addition**.

Formula

Result[i] = A[i] + B[i]

```python
a + b
```

# 2️⃣ Subtraction

## What is Subtraction?

Subtraction means finding the difference between corresponding elements of two NumPy arrays.

Like addition, NumPy performs subtraction **element-wise**.

It subtracts each element of the second array from the element at the same position in the first array.

---

## Formula

Result[i] = A[i] - B[i]

---

## Why is it Useful?

Subtraction is widely used in:

- Finding errors (Predicted - Actual)
- Data preprocessing
- Calculating differences between datasets
- Image processing (finding changes between two images)
- Machine Learning loss calculations

```python
a = np.array([10, 20, 30])
b = np.array([4, 5, 6])

a - b
```

```python
b-a
```

### Important Observation

Changing the order of arrays changes the answer.

A - B ≠ B - A

Therefore, subtraction is **not commutative**.

# 3️⃣ Multiplication

## What is Multiplication?

NumPy multiplies corresponding elements of two arrays.

This is called **Element-wise Multiplication**.

---

## Formula

Result[i] = A[i] × B[i]

---

## ⚠️ Important

Many beginners think that using `*` performs Matrix Multiplication.

This is incorrect.

In NumPy:

`*` → Element-wise Multiplication

`@` or `np.dot()` → Matrix Multiplication

```python
a = np.array([2,4,6])
b = np.array([3,5,7])

a * b
```

# 5️⃣ Axis Parameter

## What is an Axis?

An **Axis** represents the direction along which NumPy performs an operation.

Many NumPy functions such as:

- np.sum()
- np.mean()
- np.max()
- np.min()

allow us to specify an axis.

Without an axis, NumPy performs the operation on the entire array.

With an axis, NumPy performs the operation along a particular direction.

---

## Why is Axis Important?

Most real-world datasets are not one-dimensional.

They usually have rows and columns.

The **axis parameter** helps us decide whether an operation should be performed:

- Row-wise
- Column-wise
- Entire Array

```python
import numpy as np
arr = np.array([[1,2,3],
                [4,5,6]])

arr
```

```python
np.sum(arr)
```

### Explanation

Since we did not specify an axis,

NumPy adds every element of the array.

1 + 2 + 3 + 4 + 5 + 6

↓

21

```python
np.sum(arr, axis=0)
```

### What happened?

For `axis = 0`,

NumPy moves **down each column** and performs the operation.

Column 1

1 + 4 = 5

Column 2

2 + 5 = 7

Column 3

3 + 6 = 9

Result

[5,7,9]

```python
np.sum(arr, axis=1)
```

### What happened?

For `axis = 1`,

NumPy moves across each row.

Row 1

1 + 2 + 3 = 6

Row 2

4 + 5 + 6 = 15

Result

[6,15]

# 7️⃣ Broadcasting

## What is Broadcasting?

Broadcasting is a mechanism used by NumPy that allows arrays of different but compatible shapes to perform arithmetic operations.

Instead of physically copying data,

NumPy intelligently adjusts the arrays so that element-wise operations become possible.

---

## Why is Broadcasting Needed?

Suppose we have an array containing marks of students.

We want to add 5 bonus marks to every student.

Without broadcasting, we would need to write a loop.

With broadcasting, NumPy performs the operation automatically.

Broadcasting makes code:

- Shorter
- Faster
- Easier to read

```python
a = np.array([1,2,3])

a + 5
```

# Broadcasting (Part 2)

## Can NumPy Add Arrays of Different Shapes?

The answer is:

**Yes, but only if their shapes are compatible.**

Broadcasting does **not** mean that every pair of different-shaped arrays can be added.

NumPy first checks whether the shapes are compatible.

If they are compatible:

✅ Operation is performed.

Otherwise:

❌ NumPy raises a ValueError.

```python
a = np.array([1,2,3])

b = np.array([10])

a + b
```

```python
a = np.array([1,2,3])

b = np.array([10,20,30])

a+b
```

# Broadcasting Rules

When NumPy performs operations on two arrays of different shapes, it compares their shapes.

The comparison starts from the **last dimension** (rightmost side).

For every dimension:

✅ The sizes are equal

OR

✅ One of them is 1

If either condition is satisfied, NumPy can broadcast that dimension.

Otherwise, broadcasting fails and NumPy raises a ValueError.

```python
a = np.array([1,2,3])

b = np.array([10,20,30])

a+b
```

```python
a = np.array([1,2,3])

b = np.array([10])

a+b
```

```python
import numpy as np

a = np.array([[1,2,3],
              [4,5,6]])

b = np.array([10,20,30])

print(a+b)
```

```python
a = np.array([[1],
              [2],
              [3]])

b = np.array([10,20,30])

print(a+b)
```

## Finding the Output Shape

After confirming that broadcasting is possible:

For each dimension:

- If both sizes are equal → Result keeps that size.
- If one size is 1 → Result takes the larger size.

Examples:

3 and 3 → 3

1 and 5 → 5

2 and 1 → 2
