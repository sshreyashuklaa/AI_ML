# 1️⃣ Arithmetic Operations

## What are Arithmetic Operations?

Arithmetic operations are basic mathematical operations such as addition, subtraction, multiplication, division, modulus, floor division, and exponentiation.

Unlike Python lists, NumPy performs these operations **element-wise**.

This means that the operation is performed on corresponding elements of two arrays.

---

## Why is this Important?

Imagine you have a dataset containing one million values.

Using Python loops, performing operations on every value would take much longer.

NumPy performs these operations internally using optimized C code, making them significantly faster.

---

## Real-Life Example

Suppose you have marks of students in two tests.

Test 1

[80, 75, 90]

Test 2

[10, 15, 5]

If bonus marks need to be added:

80 + 10

75 + 15

90 + 5

NumPy performs this automatically without writing loops.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Array A :", a)
print("Array B :", b)
```

## Addition

Addition combines corresponding elements from both arrays.

This is called **Element-wise Addition**.

Formula

Result[i] = A[i] + B[i]

```python
a + b
```

# 2️⃣ Subtraction

## What is Subtraction?

Subtraction means finding the difference between corresponding elements of two NumPy arrays.

Like addition, NumPy performs subtraction **element-wise**.

It subtracts each element of the second array from the element at the same position in the first array.

---

## Formula

Result[i] = A[i] - B[i]

---

## Why is it Useful?

Subtraction is widely used in:

- Finding errors (Predicted - Actual)
- Data preprocessing
- Calculating differences between datasets
- Image processing (finding changes between two images)
- Machine Learning loss calculations

```python
a = np.array([10, 20, 30])
b = np.array([4, 5, 6])

a - b
```

```python
b-a
```

### Important Observation

Changing the order of arrays changes the answer.

A - B ≠ B - A

Therefore, subtraction is **not commutative**.

# 3️⃣ Multiplication

## What is Multiplication?

NumPy multiplies corresponding elements of two arrays.

This is called **Element-wise Multiplication**.

---

## Formula

Result[i] = A[i] × B[i]

---

## ⚠️ Important

Many beginners think that using `*` performs Matrix Multiplication.

This is incorrect.

In NumPy:

`*` → Element-wise Multiplication

`@` or `np.dot()` → Matrix Multiplication

```python
a = np.array([2,4,6])
b = np.array([3,5,7])

a * b
```

# 4️⃣ Division

NumPy divides corresponding elements.

Formula

Result[i] = A[i] / B[i]

The result of `/` is usually a floating-point number, even if both arrays contain integers.

```python
a = np.array([10,20,30])
b = np.array([2,5,6])

a / b
```

# 5️⃣ Floor Division

Floor Division returns the quotient after removing the decimal part.

Operator

//

```python
a = np.array([10,20,30])

b = np.array([3,6,7])

a // b
```

# 6️⃣ Modulus

Modulus returns the remainder after division.

Operator

%

```python
a = np.array([10,20,30])

b = np.array([3,6,7])

a % b
```

# 7️⃣ Exponentiation

Exponentiation raises every element of an array to the specified power.

Operator

**

```python
a = np.array([2,3,4])

a ** 2
```

# 2️⃣ Scalar Operations

## What is a Scalar?

A **Scalar** is a single numerical value.

Examples:

5
10
-3
0
2.5

Unlike arrays, a scalar contains only one value.

---

## What are Scalar Operations?

Scalar operations mean performing an arithmetic operation between:

• One NumPy Array

and

• One Single Number (Scalar)

NumPy automatically applies the operation to **every element** of the array.

No loop is required.

```python
a = np.array([1,2,3])

a
```

```python
a + 5
```

### Explanation

NumPy automatically applies the scalar value to every element of the array.

Internally, NumPy performs the operation on each element.

Conceptually:

1 + 5

2 + 5

3 + 5

↓

[6,7,8]

The important thing is that **you do not have to write the loop yourself**.

```python
a - 2
```

```python
a * 10
```

```python
a / 2
```

```python
a ** 3
```
