# Topic 1 - Scalar

## What is a Scalar?

A scalar is a **single numerical value**.

It has:
- No direction
- No rows or columns
- No dimensions to navigate

Examples:
- 5
- -10
- 3.14
- 100

A scalar represents only one quantity.

In NumPy, a scalar is represented as a **0-dimensional array**.

```python
import numpy as np

scalar = np.array(10)

print(scalar)
```

```python
print(type(scalar))
```

```python
print(scalar.ndim)
```

```python
print("Shape :", scalar.shape)
print("Size  :", scalar.size)
print("Datatype :", scalar.dtype)
```

# Topic 2 - Vector

## What is a Vector?

A vector is a collection of values arranged along a **single axis**.

Unlike a scalar, a vector stores multiple values.

Examples:

[10, 20, 30]

[5, 8, 12, 15]

Characteristics

- Has one dimension
- Stores multiple values
- Shape = (n,)
- ndim = 1

In NumPy, a vector is simply a **1D array**.

```python
import numpy as np

vector = np.array([10, 20, 30, 40, 50])

print(vector)
```

```python
print(type(vector))
print(vector.ndim)
print(vector.shape)
print(vector.size)
print(vector.dtype)
```

```python
print(vector[0])

print(vector[2])

print(vector[-1])
```

```python
vector[1] = 100

print(vector)
```

```python
arr = np.array([[10, 20, 30]])
np.shape(arr)
```

# Topic 3 - Matrix

## What is a Matrix?

A matrix is a collection of values arranged in **rows and columns**.

Unlike a vector, a matrix has **two dimensions**.

Examples

[
 [1, 2, 3],
 [4, 5, 6]
]

Characteristics

- Two-dimensional
- Has rows and columns
- Shape = (rows, columns)
- ndim = 2

In NumPy, a matrix is simply a **2D array**.

```python
import numpy as np

matrix = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(matrix)
```

```python
print(type(matrix))
print(matrix.ndim)
print(matrix.shape)
print(matrix.size)
print(matrix.dtype)
```

```python
print(matrix[0])
```

```python
print(matrix[1,2])
```

```python
matrix[0,1] = 100

print(matrix)
```
