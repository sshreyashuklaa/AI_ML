# Indexing in NumPy

## What is Indexing?

Indexing is the process of accessing a specific element of an array using its position (index).

Python follows **0-based indexing**, meaning the first element has index 0.

First index = 0

Last Index = Length − 1

Syntax: array[index]

Invalid index gives IndexError.

Example:

Array:
[10, 20, 30, 40, 50]

Index:
 0   1   2   3   4

```python
import numpy as np

arr = np.array([10,20,30,40,50])

print(arr)

print(arr[0])
```

# Negative Indexing

Negative indexing allows us to access elements from the end of an array.

Instead of counting from the beginning, Python counts from the last element.

Last element → -1, Second last → -2, First element → -Length

Syntax: array[-index]

Index Representation

Positive Index

 0   1   2   3   4
10  20  30  40  50

Negative Index

-5  -4  -3  -2  -1
10  20  30  40  50

Negative indexing is useful when we want to access elements from the end without knowing the array length.

```python
import numpy as np

arr = np.array([10,20,30,40,50])

arr
```

```python
print(arr[-1])
```

# Slicing in NumPy

## What is Slicing?

Slicing is used to access multiple consecutive elements from an array.

Instead of retrieving a single element, slicing returns a portion (sub-array) of the original array.

### Syntax

array[start : stop : step]

where,

- start → Starting index (included)
- stop → Ending index (excluded)
- step → Number of elements to skip (default = 1)

Important Rule:

Start is included.
Stop is excluded.

```python
import numpy as np

arr = np.array([10,20,30,40,50])

arr
```

```python
arr[1:4]
```

# Step Slicing

## What is Step Slicing?

Step slicing allows us to skip elements while creating a slice.

### Syntax

array[start : stop : step]

where,

- start → Starting index (included)
- stop → Ending index (excluded)
- step → Number of positions to move after selecting an element

Default value of step is 1.

Positive step moves from left to right.

Negative step moves from right to left.

```python
import numpy as np

arr = np.array([10,20,30,40,50,60,70,80])

arr
```

```python
arr[0:8:1]
```

```python
arr[0:8:2]
```

```python
arr[1:8:2]
```

```python
arr[2:7:3]
```

```python
arr[::2]
```

```python
arr[1::2]
```

# Negative Step Slicing

## What is Negative Step?

A negative step allows us to traverse an array from right to left.

Instead of moving forward, Python moves backward.

### Syntax

array[start : stop : step]

where,

- start → Starting index
- stop → Ending index (excluded)
- step → Negative value (moves backward)

Common Examples

[::-1] → Reverse the array

[::-2] → Every second element from the end

```python
import numpy as np

arr = np.array([10,20,30,40,50,60,70,80])

arr
```

```python
arr[::-1]
```

```python
arr[::-2]
```

```python
arr[6:1:-1]
```

```python
arr[5:0:-2]
```

# 2D Array Indexing

## What is a 2D Array?

A 2D array is an array of arrays.

It consists of rows and columns and is commonly used to represent tables, matrices, images, and datasets.

### Syntax

array[row_index, column_index]

Important:

- First index → Row
- Second index → Column

```python
import numpy as np

arr = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

arr
```

```python
print(arr[0,0])
```

```python
arr[1,2]
```

```python
arr[1]
```

# 2D Array Slicing

## What is 2D Slicing?

2D slicing is used to access multiple rows and multiple columns from a 2D array.

Unlike 1D slicing, we specify slicing separately for rows and columns.

### Syntax

array[row_slice, column_slice]

where,

- row_slice → Selects rows
- column_slice → Selects columns

General Form

array[row_start : row_stop : row_step,
      col_start : col_stop : col_step]

```python
import numpy as np

arr = np.array([
    [11,12,13,14],
    [21,22,23,24],
    [31,32,33,34],
    [41,42,43,44]
])

arr
```

```python
arr[1:3]
```

```python
arr[:,1]
```

```python
arr[:,:]
```

```python
arr[2,:]
```

```python
arr[:,1:3]
```

```python
arr[1:3,1:3]
```

```python
arr[1:3,2]
```

# Advanced 2D Slicing

## What is Advanced 2D Slicing?

Advanced 2D slicing allows us to select rows and columns using step values and reverse directions.

It helps us efficiently extract patterns from datasets without using loops.

### Syntax

array[row_start : row_stop : row_step,
      col_start : col_stop : col_step]

Remember:

- First part → Rows
- Second part → Columns
- Step controls how many elements are skipped.
- Negative step reverses the direction.

```python
import numpy as np

arr = np.array([
    [11,12,13,14],
    [21,22,23,24],
    [31,32,33,34],
    [41,42,43,44]
])

arr
```

```python
arr[::2, :]
```

```python
arr[:, ::2]
```

```python
arr[::2, ::2]
```

```python
arr[::-1, :]
```

```python
arr[:, ::-1]
```

```python
arr[::-1, ::-1]
```

# Boolean Indexing (Filtering)

## What is Boolean Indexing?

Boolean Indexing is a technique used to filter elements from an array based on a condition.

Instead of accessing elements using positions (indices), we access them using True and False values.

### Why is it Important?

- Data Filtering
- Data Cleaning
- Machine Learning
- Data Analysis
- Feature Selection

Boolean Indexing is one of the most frequently used NumPy concepts.

```python
import numpy as np

arr = np.array([10,20,30,40,50])

print(arr)
```

```python
arr>25
```

```python
arr[arr>25]
```

# Multiple Conditions in Boolean Indexing

## What are Multiple Conditions?

Sometimes, we need to filter data using more than one condition.

Examples:
- Marks greater than 40 AND less than 80
- Age greater than 18 OR Salary greater than 50000

NumPy provides special operators to combine multiple conditions.

### Operators

&  → AND

|  → OR

~  → NOT

⚠ Never use Python's `and`, `or`, and `not` with NumPy arrays.

```python
import numpy as np

marks = np.array([25, 42, 58, 76, 89, 91, 35])
```

```python
(marks > 40) & (marks < 90)
```

```python
marks[(marks > 40) & (marks < 90)]
```

```python
marks[(marks < 40) | (marks > 90)]
```

```python
marks[~(marks >= 40)]
```

```python
arr = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])
```

```python
arr[arr > 50]
```

# Fancy Indexing

## What is Fancy Indexing?

Fancy Indexing is a technique used to access multiple specific elements, rows, or columns at once using lists or arrays of indices.

Unlike slicing, Fancy Indexing allows us to choose any positions in any order.

### Why is it useful?

- Selecting specific samples
- Rearranging data
- Feature selection
- Machine Learning preprocessing

```python
import numpy as np

arr = np.array([10,20,30,40,50])
```

```python
arr[[0,3,4]]
```

```python
arr[[1,1,3]]
```

```python
arr = np.array([
    [11,12,13],
    [21,22,23],
    [31,32,33],
    [41,42,43]
])
```

```python
arr[[0,2,3]]
```

```python
arr[[3,1,0]]
```

```python
arr[[-1,-3]]
```
