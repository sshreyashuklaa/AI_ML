# Universal Functions (ufuncs)

A Universal Function (ufunc) is a function that performs
element-wise operations on NumPy arrays.

Examples

np.add()
np.subtract()
np.multiply()
np.divide()
np.sqrt()
np.power()

Why use ufuncs?

• Fast
• Memory efficient
• Works directly on arrays
• Implemented in optimized C code

```python
import numpy as np

marks = np.array([80,75,92,60,88])

marks + 5
```

```python
a = np.array([1,2,3])

b = np.array([10,20,30])

a + b
```

```python
a = np.array([2,4,6])
b = np.array([1,3,5])

print(a + b)
```

# np.median()

Returns the middle value of the sorted data.

Mean ≠ Median

Mean
-----
Average of all values.

Median
-------
Middle value after sorting.

Median is less affected by extreme values (outliers).

```python
import numpy as np

arr = np.array([20,22,21,23,500])

np.median(arr)
```

```python
arr = np.array([10,20,30,40])
```

```python
np.median([10,20,30,40])
```

```python
arr = np.array([[1,2],
                [3,4]])

np.median(arr)
```

```python
arr = np.array([[1,5],
                [3,7],
                [9,2]])

np.median(arr, axis=0)
```

```python
np.median(arr, axis=1)
```

# Standard Deviation and Variance

Variance
--------
Measures how spread out the data is.

Standard Deviation
------------------
Square root of the variance.

Small std → Data is close to the mean.
Large std → Data is spread out.

```python
import numpy as np

arr = np.array([10,20,30,40,50])

np.var(arr)
```

```python
np.std(arr)
```

```python
a = np.array([49,50,51,50,50])

print(np.std(a))
```

```python
b = np.array([10,90,20,80,50])

print(np.std(b))
```

```python
arr = np.array([[1,2],
                [3,4]])
```

```python
np.std(arr)
```

```python
np.std(arr, axis=0)
```

```python
np.std(arr, axis=1)
```

# Product and Cumulative Functions

np.prod()
----------
Returns the product of all elements.
Think of it as the multiplication version of sum().

np.cumsum()
------------
Returns cumulative (running) sum.

np.cumprod()
-------------
Returns cumulative (running) product.

```python
import numpy as np

arr = np.array([2,3,4])

np.prod(arr)
```

```python
arr = np.array([[1,2],
                [3,4]])

np.prod(arr)
```

```python
np.prod(arr, axis=0)
```

```python
np.prod(arr, axis=1)
```

```python
arr = np.array([10,20,30,40])

np.cumsum(arr)
```

```python
arr = np.array([2,3,4])

np.cumprod(arr)
```
