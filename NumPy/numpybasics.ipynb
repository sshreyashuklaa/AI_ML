# NumPy Fundamentals

NumPy (Numerical Python) is the fundamental library used for numerical computing in Python.

It provides:

- Fast multidimensional arrays
- Mathematical operations
- Matrix computations
- Linear algebra
- Random number generation
- Foundation for Machine Learning, Data Science and Deep Learning

Almost every AI library (Pandas, Scikit-Learn, TensorFlow, PyTorch) internally uses NumPy.

## Why not Python Lists?

Python lists are flexible but slow for numerical computation.

NumPy arrays:

- consume less memory
- execute faster
- support vectorized operations
- are optimized using C

## Creating a NumPy Array

The np.array() function is used to create a NumPy array from Python lists.

Syntax:

np.array(iterable)

```python
import numpy as np
arr = np.array([10,20,20,40])
print(arr)
```

```python
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([1.5,2.5,3.5])
arr3 = np.array(["Apple, Banana, Mango"])
arr4 = np.array([True, False, True])

print(arr1)
print(arr2)
print(arr3)
print(arr4)
```

# Array Attributes

Every NumPy array stores metadata (information about the array).

Array attributes help us understand:

- Number of dimensions
- Shape of the array
- Total number of elements
- Data type of elements
- Memory occupied by each element
- Total memory occupied by the array

These attributes are useful for debugging, data analysis, and Machine Learning.

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60]])

print(arr)
```

## ndim

`ndim` returns the number of dimensions (axes) of an array.

Examples:

- 1D array → ndim = 1
- 2D array → ndim = 2
- 3D array → ndim = 3

```python
print(arr.ndim)
```

## shape

`shape` returns the size of each dimension.

Output is returned as a tuple.

For a 2D array:

(rows, columns)

```python
print(arr.shape)
```

## size

`size` returns the total number of elements present in the array.

```python
print(arr.size)
```

## dtype

`dtype` returns the data type of the elements stored in the array.

```python
print(arr.dtype)
```

## itemsize

`itemsize` returns the memory occupied by one element (in bytes).

```python
print(arr.itemsize)
```

## nbytes

`nbytes` returns the total memory occupied by the array.

Formula:

Total Memory = size × itemsize

```python
print(arr.nbytes)
```

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60]])

print("Dimensions :", arr.ndim)
print("Shape      :", arr.shape)
print("Size       :", arr.size)
print("Data Type  :", arr.dtype)
print("Item Size  :", arr.itemsize)
print("Total Bytes:", arr.nbytes)
```
